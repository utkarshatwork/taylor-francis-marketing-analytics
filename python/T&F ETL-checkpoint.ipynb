<h2 style="color: #1e40af; border-bottom: 2px solid #e2e8f0; padding-bottom: 8px; margin-top: 40px; font-weight: bold;">Phase 1: Data Cleaning & Transformation</h2>

**Step 1: Data Exploration & Initial Health Check**

Before we start cleaning or analyzing anything, we need to understand the current state of our raw data.
In this first step, I am loading the authors.csv file to do a quick "health check." I want to see how much data we have and look out for common issues like missing values, inconsistent texts and like typos, or illogical numbers that don't make sense. Finding these issues now will make our cleaning process much more targeted and ensure our final dashboard is accurate.

```python
import pandas as pd

# 1. Load the raw authors dataset
authors_df = pd.read_csv("authors.csv")

# 2. Check the size of the dataset (rows and columns)
print(f"Dataset Size: {authors_df.shape[0]} rows and {authors_df.shape[1]} columns\n")

# 3. Look for missing data and check data types
print("--- Column Info & Missing Values ---")
authors_df.info()

# 4. Pull a random sample of 5 rows to eyeball the data
print("\n--- Random Data Sample ---")
authors_df.sample(30)
```

**Step 2: Data Transformation (Cleaning the Authors Table)**

Based on the initial health check, I identified three specific issues that need to be resolved before this data is usable:

Missing Data: About 5% of authors are missing their institution_type. Dropping these rows would lose valuable revenue data later, so I will impute them with "Unknown".

Inconsistent Job Titles: There are several messy formatting variations of "Pharma Research Scientist" (e.g., casing issues, abbreviations). I will standardize these to ensure our author segmentation is accurate.

Numerical Outliers: A small number of records contain impossible historical_publications counts (like -1 or 9999). I'll cap these at realistic biological limits so they don't skew our averages.

```python
# 1. Fill missing institution types with 'Unknown'
authors_df['institution_type'] = authors_df['institution_type'].fillna('Unknown')

# 2. Standardize the messy job titles
# Create a mapping dictionary for the typos we found
title_fixes = {
    "pharma scientist": "Pharma Research Scientist",
    "PHARMA RESEARCHER": "Pharma Research Scientist",
    "Research Sci.": "Pharma Research Scientist",
    "pharma research scientist": "Pharma Research Scientist"
}

# Remove hidden spaces at the edges, then apply our fixes
authors_df['job_title'] = authors_df['job_title'].str.strip().replace(title_fixes)

# 3. Handle the impossible numerical outliers
# If a publication count is less than 0, set it to 0
authors_df.loc[authors_df['historical_publications'] < 0, 'historical_publications'] = 0

# If a publication count is over 100 (e.g., 999), cap it at a realistic maximum of 50
authors_df.loc[authors_df['historical_publications'] > 100, 'historical_publications'] = 50

# --- VERIFICATION ---
# Let's prove the data is clean now
print("--- Missing Values After Cleaning ---")
print(authors_df.isnull().sum())

print("\n--- Cleaned Job Title Counts ---")
print(authors_df['job_title'].value_counts())
```

**Step 3: Data Exploration - Marketing Interactions**

Next, I am extracting the marketing_interactions.csv file. This table is the "connective tissue" of the project because it links author behavior to our specific marketing campaigns.

Before cleaning, I need to inspect the data for two main issues:

Channel Consistency: I need to ensure the channel names here perfectly match our marketing_spend table. If there are typos like " Social Ads" or lowercase "email", our SQL joins will fail later and our ROAS calculations will be inaccurate.

Date Variability: I need to check if the interaction_date column uses a consistent format. Mixed formats (like using both slashes and hyphens) will prevent us from sorting the author journey chronologically for the multi-touch attribution model.

```python
# 1. Load the raw marketing interactions
interactions_df = pd.read_csv("marketing_interactions.csv")

# 2. Inspect the 'channel' column for consistency issues
print("--- Marketing Channel Groupings (Raw) ---")
print(interactions_df['channel'].value_counts())

# 3. Inspect date formatting (looking for mixed DD-MM and MM/DD formats)
print("\n--- Sample of Interaction Dates ---")
display(interactions_df[['interaction_date', 'channel']].sample(10))
```

**Step 4: Data Transformation - Standardizing Interactions**

Based on the exploration, two critical transformations are required for the marketing_interactions table:

String Standardization: I need to strip/remove trailing whitespaces and normalize the casing in the channel column. This ensures precise JOIN operations with our marketing_spend table later in SQL.

Chronological Harmonization: The interaction_date column contains a mix of UK (DD-MM-YYYY) and US (MM/DD/YYYY) formats. I am implementing a custom parsing function to standardize these into a unified datetime object (YYYY-MM-DD). This is a non-negotiable step for the Multi-Touch Attribution model, which relies entirely on strict chronological sorting.

```python
# 1. Standardize the 'channel' strings
# Strip invisible spaces from the edges and capitalize the first letter of each word
interactions_df['channel'] = interactions_df['channel'].str.strip().str.title()

# 2. Harmonize the mixed date formats
def parse_interaction_date(date_str):
    """Custom function to handle mixed UK and US date formats."""
    try:
        # Try to read it as the standard UK format (Day-Month-Year)
        dt = pd.to_datetime(date_str, format='%d-%m-%Y')
    except ValueError:
        # If it crashes (Value Error), fall back to the US format (Month/Day/Year)
        dt = pd.to_datetime(date_str, format='%m/%d/%Y')

        # Force the output to be a clean string for CSV export
    return dt.strftime('%Y-%m-%d')

# Apply our custom function to the date column
interactions_df['interaction_date'] = interactions_df['interaction_date'].apply(parse_interaction_date)

# --- VERIFICATION ---
print("--- Cleaned Marketing Channels ---")
print(interactions_df['channel'].value_counts())

print("\n--- Verified Date Formatting (YYYY-MM-DD) ---")
display(interactions_df[['interaction_date', 'channel']].sample(5))
```

<h2 style="color: #1e40af; border-bottom: 2px solid #e2e8f0; padding-bottom: 8px; margin-top: 40px; font-weight: bold;">Phase 3: Audience Segmentation using Cluster Analysis</h2>

**Step 1: Feature Engineering & Data Scaling**

Having identified Corporate users as our primary target for highly optimized Email campaigns, the next challenge is preventing list fatigue. Sending high-frequency "Call for Papers" emails to the entire Corporate database will inevitably lead to high unsubscribe rates among casual readers.

To solve this, I am utilizing K-Means Cluster Analysis to segment the Corporate database based on their actual engagement behavior, rather than their static job titles.

Before feeding data into the clustering algorithm, I must engineer a "behavioral profile" for each user. I am aggregating the raw interaction logs to calculate the total volume of Email and Webinar engagements per user. Finally, I will apply a StandardScaler to normalize the variance, ensuring the algorithm weights both channels equally during the distance calculations.

```python
# Import Machine Learning libraries
import pandas as pd
from sklearn.preprocessing import StandardScaler

# 1. Load the core datasets

# 2. Filter the universe: We ONLY want to analyze Corporate users
corporate_authors = authors_df[authors_df['institution_type'] == 'Corporate']['author_id']
corp_interactions = interactions_df[interactions_df['author_id'].isin(corporate_authors)]

# 3. Feature Engineering: Create a behavioral profile for each user
# Pivot the data so every row is one Author, and columns are their total clicks per channel
behavioral_df = pd.pivot_table(
    corp_interactions, 
    index='author_id', 
    columns='channel', 
    aggfunc='size', 
    fill_value=0
)

# Isolate the exact behavioral features we want the algorithm to analyze
# (We are focusing on our top two engagement channels)
features = ['Email', 'Webinar Series']
X = behavioral_df[features].copy()

# 4. Data Scaling: Normalize the numerical variance 
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 5. Preview the scaled data alongside the original counts
# We attach the scaled numbers back to the dataframe for easy viewing
behavioral_df['Email_Scaled'] = X_scaled[:, 0]
behavioral_df['Webinar_Scaled'] = X_scaled[:, 1]

# Display the top 5 rows to verify the transformation
behavioral_df[['Email', 'Webinar Series', 'Email_Scaled', 'Webinar_Scaled']].head()
```

**Step 2: Determining Optimal Clusters (The Elbow Method)**

To ensure our segmentation is mathematically sound and not arbitrary, I am utilizing the Elbow Method to determine the optimal number of clusters (k). By plotting the Within-Cluster Sum of Squares (WCSS) against the number of clusters, we can identify the "elbow" point. This point represents the optimal balance between high cohesion within clusters and avoiding unnecessary model complexity.

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import os

# 1. Suppress the Windows MKL memory leak warning for a clean notebook
os.environ['OMP_NUM_THREADS'] = '1'
warnings.filterwarnings('ignore')

# 2. Calculate WCSS (Within-Cluster Sum of Squares) for k=1 to k=10
wcss = []
k_range = range(1, 11)

for k in k_range:
    # random_state=42 for reproducibility, n_init=10 as best practice
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

# 3. Visualize the Elbow Curve
plt.figure(figsize=(10, 6))
sns.lineplot(x=k_range, y=wcss, marker='o', color='#2a9d8f', linewidth=2, markersize=8)

# 4. Format the plot aesthetics
plt.title('Elbow Method: Determining Optimal Audience Segments', fontsize=14, fontweight='bold')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.xticks(k_range)
plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

**Step 3: K-Means Model Execution & Cluster Profiling**

Based on the Elbow Method, the optimal number of behavioral segments is k=3.

I am now initializing the K-Means algorithm to assign every Corporate user into one of these three distinct behavioral clusters. Once the model has segmented the audience, I will aggregate the original interaction volumes to profile the distinct characteristics of each cluster and identify our highest-value target segment.

```python
# 1. Initialize the K-Means model with our optimal number of clusters (k=3)
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)

# 2. Fit the model to our scaled data and assign a cluster label to every user
behavioral_df['Cluster'] = kmeans.fit_predict(X_scaled)

# 3. Profile the Clusters
# We group the data by the new 'Cluster' label and calculate the average interactions
cluster_profile = behavioral_df.groupby('Cluster')[['Email', 'Webinar Series']].mean().round(2)

# Count how many unique authors landed in each cluster
cluster_profile['Total_Users'] = behavioral_df.groupby('Cluster').size()

# Rename the columns for a clean, professional presentation
cluster_profile = cluster_profile.rename(columns={
    'Email': 'Avg Email Opens',
    'Webinar Series': 'Avg Webinars Attended'
})

# Display the final summary table
cluster_profile
```

**Step 4: Segment Validation & Demographic Reveal**

Having isolated Cluster 1 as our high-intent engagement segment (averaging 13.48 email opens per user), the final step is to validate the demographic makeup of this group.

To ensure our marketing spend is directed efficiently, I am merging the algorithm's behavioral cluster labels back with our original demographic data from authors table. This will definitively prove who these hyper-engaged users are, allowing us to tailor our Call-for-Papers (CFP) messaging directly to their specific persona.

```python
# 1. Merge the new Machine Learning labels back with the original Author data
final_df = pd.merge(behavioral_df.reset_index(), authors_df, on='author_id')

# 2. Isolate ONLY our High-Intent "Whale" group (Whale Cluster)
target_cluster_id = cluster_profile['Avg Email Opens'].idxmax()
whale_cluster = final_df[final_df['Cluster'] == target_cluster_id]

# 3. Calculate the percentage breakdown of job titles inside this VIP cluster
whale_demographics = whale_cluster['job_title'].value_counts()
whale_percentages = (whale_demographics / len(whale_cluster) * 100).round(1)

# 4. Display the demographic validation
print("💎 DEMOGRAPHIC BREAKDOWN OF THE HIGH-INTENT SEGMENT (Whale Cluster) 💎\n")
for title, percentage in whale_percentages.items():
    print(f"- {title}: {percentage}%")
```
